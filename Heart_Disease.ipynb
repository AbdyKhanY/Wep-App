Here is a simple explanation of each variable:

1. **Age**: How old the person is.
2. **Sex**: Whether the person is a boy or a girl.
3. **Chest Pain Type**: Different types of chest pain a person can feel.
4. **Resting Blood Pressure**: The pressure of the blood in the person’s veins when they are resting.
5. **Serum Cholesterol in mg/dl**: The amount of fat in the person’s blood.
6. **Fasting Blood Sugar > 120 mg/dl**: If the person's blood sugar is high after not eating for a while.
7. **Resting Electrocardiographic Results**: How the heart's electrical activity looks when the person is not moving.
8. **Maximum Heart Rate Achieved**: The highest heart rate the person reaches during exercise.
9. **Exercise Induced Angina**: Chest pain that happens when the person is exercising.
10. **Oldpeak**: How much the ST segment in the heart's activity drops after exercise.
11. **The Slope of the Peak Exercise ST Segment**: The slope shape of the heart's ST segment after exercise.
12. **Number of Major Vessels (0-3) Colored by Fluoroscopy**: How many major blood vessels are seen using a special camera.
13. **Thal**: Whether the person's heart is normal, has a fixed defect, or a reversible defect.
14. **Target**: 0- if a person is healthy and 1 if the person has a heart disease.

# MACHINE LEARNING

```python
df.head()
```

**TRAINING THE MODEL**

```python
from sklearn.model_selection import train_test_split

#choosing the X and Y variables
x= df.drop('target',axis=1)
y=df['target']

# Splitting the data into training and testing sets
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2, random_state=2)
print("Training: " , train_x.shape, train_y.shape)
print("Testing: " , test_x.shape, test_y.shape)
```

```python
import warnings # to filter and remove warnings
warnings.filterwarnings("ignore")
```

**Logistic Regression Model Training**

```python
from sklearn.linear_model import LogisticRegression

#fitting the model
model=LogisticRegression()
model.fit(train_x,train_y)
```

**MODEL SCORE**

```python
#Accuracy on training data
from sklearn.metrics import accuracy_score

train_x_prediction=model.predict(train_x)
train_accuracy=accuracy_score(train_x_prediction,train_y)
print("Accuracy on training data : ", train_accuracy)
```

```python
#Accuracy on Test data

test_x_prediction=model.predict(test_x)
test_accuracy=accuracy_score(test_x_prediction,test_y)
print("Accuracy on test data : ", test_accuracy)
```

**FEATURE IMPORTANCE**

```python
# Access the coefficients
coefficients = model.coef_[0]

# Create a DataFrame to display feature names and their corresponding coefficients
feature_importance = pd.DataFrame({'Feature': train_x.columns, 'Coefficient': coefficients})

# Sort the DataFrame by absolute values of coefficients in descending order
feature_importance['Absolute_Coefficient'] = feature_importance['Coefficient'].abs()
feature_importance = feature_importance.sort_values(by='Absolute_Coefficient', ascending=False).drop('Absolute_Coefficient', axis=1)

# Reset the index
feature_importance = feature_importance.reset_index(drop=True)

# Display the feature importance ranking
print(feature_importance)
```

**PLOTTING FEATURE IMPORTANCE**

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Feature', y='Coefficient', data=feature_importance, palette = 'viridis')
plt.title('Feature Importance')
plt.xlabel('Feature')
plt.ylabel('Coefficient')

# Add data labels above the bars
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:.2f}', (p.get_x() + p.get_width() / 2., height),
                ha='center', va='bottom', xytext=(0, 5), textcoords='offset points')
plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score

# Assuming 'test_y' are the true labels and 'predictions' are the model's predictions
# Let's also assume 'model' is your logistic regression model

# Perform predictions
predictions = model.predict(test_x)

# Calculate probabilities for ROC-AUC
probabilities = model.predict_proba(test_x)[:, 1]

# Calculate metrics
precision = precision_score(test_y, predictions)
recall = recall_score(test_y, predictions)
f1 = f1_score(test_y, predictions)
roc_auc = roc_auc_score(test_y, probabilities)

print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f} ')
print(f'F1-Score: {f1:.4f}')
print(f'ROC-AUC: {roc_auc:.4f}')
```

**Precision score** ▶ A precision score of 0.8286 means that out of all the instances the model predicted as positive, approximately 82.86% were actually true positives.

**Recall**  ▶ A recall score of 1.00 indicates that the model correctly identifies all actual positive instances in the dataset, without missing any.

**F1-Score** ▶ The F1-score of 0.9062 indicates a strong balance between precision and recall, reflecting the model's overall performance in binary classification tasks.

**Roc-AuC** ▶ The ROC-AUC score of 0.9515 suggests that the model has excellent discriminatory ability in distinguishing between positive and negative classes.
